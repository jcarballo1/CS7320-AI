## Tree structure

Here is an implementation of the basic node structure for the search algorithms (see Fig 3.7 on page 73). I have added a method that extracts the path from the root node to the current node. It can be used to get the path when the search is completed.

```python
class Node:
    def __init__(self, pos, parent, action, cost):
        self.pos = tuple(pos)    # the state; positions are (row,col)
        self.parent = parent     # reference to parent node. None means root node.
        self.action = action     # action used in the transition function (root node has None)
        self.cost = cost         # for uniform cost this is the depth. It is also g(n) for A* search

    def __str__(self):
        return f"Node - pos = {self.pos}; action = {self.action}; cost = {self.cost}"
    
    def get_path_from_root(self):
        """returns nodes on the path from the root to the current node."""
        node = self
        path = [node]
    
        while not node.parent is None:
            node = node.parent
            path.append(node)
        
        path.reverse()
        
        return(path)
```

## Task 1: Defining the search Problem [1 point]

Define the components of the search problem:

* Initial state
* Actions
* Transition model
* Goal state
* Path cost

- Initial State
    - The state in which the machine should start; In this case, starting location S in the maze
- Actions
    - Set of all actions possible by the agent; north, south, east, and west
- Transition Model
    - The state after performing an action; location after moving n,s,e, or w
- Goal State
    - Destination state of the agent; Ending location G
- Path Cost
    - Sum of all the action costs; Each action has a cost of 1

## Task 2: Breadth-first and Depth-first [4 points]

Implement these search strategies. Follow the pseudocode in the textbook/slides. You can use the tree structure shown above to extract the final path from your solution.

__Notes:__
* You can find maze solving implementations online that use the map to store information. While this is an effective idea for this two-dimensional navigation problem, it typically cannot be used for other search problems. Therefore, follow the textbook and only store information during search in the tree, and the `reached` and `frontier` data structures.
* DSF can be implemented using the BFS tree search algorithm and changing the order in which the frontier is expanded (this is equivalent to best-first search with path length as the criterion to expand the next node). However, to take advantage of the significantly smaller memory footprint, you need to implement DFS in a different way without a `reached` data structure and by releasing nodes that are not needed anymore. 
* If DFS does not use a `reached` data structure, then its cycle checking abilities are limited. Remember, that DSF is incomplete if cycles cannot be prevented. You will see in your experiments that open spaces are a problem.

```python
from collections import deque

def get_actions(maze, node):
    actions = []
    
    # store actions for easy access
    north = node.pos[0] - 1, node.pos[1]
    south = node.pos[0] + 1, node.pos[1]
    east = node.pos[0], node.pos[1] + 1
    west = node.pos[0], node.pos[1] - 1
    
    if look(maze, north) != "X":
        action = Node(pos = north, parent = node, action = "N", cost = node.cost + 1)
        actions.append(action)
    if look(maze, south) != "X":
        action = Node(pos = south, parent = node, action = "S", cost = node.cost + 1)
        actions.append(action)
    if look(maze, east) != "X":
        action = Node(pos = east, parent = node, action = "E", cost = node.cost + 1)
        actions.append(action)
    if look(maze, west) != "X":
        action = Node(pos = west, parent = node, action = "W", cost = node.cost + 1)
        actions.append(action)
        
    return actions

# update maze for show_maze
def update_maze(maze, path):
    for node in path:
        pos = node.pos
        if(look(maze, pos) != "S" and look(maze, pos) != "G"):
            maze[pos] = "P"
            
def bfs(maze):
    # store positions of initial and goal states
    start = find_pos(maze, what = "S")
    end = find_pos(maze, what = "G")
    # create node at position of initial state
    root_node = Node(pos = start, parent = None, action = None, cost = 0)
    # create doubly ended queue with only root
    frontier = deque()
    frontier.append(root_node)
    # create empty set for explored nodes
    explored = {}
    nodes_expanded = 0
    max_t_depth = 0
    max_f_size = 0
    
    #while frontier is not empty
    while frontier:
        # gets and removes first element in queue
        curr_node = frontier.popleft()
        # checks if node is in goal state position; if so return path
        if look(maze, curr_node.pos) == "G":
            path = curr_node.get_path_from_root()
            update_maze(maze, path)
            return path, [n.cost for n in path][-1], nodes_expanded, max_t_depth, max_f_size
        
        # add curr_node to explored set
        explored[curr_node.pos] = curr_node
        # get action/children nodes around curr_node
        actions = get_actions(maze, curr_node)
        nodes_expanded = nodes_expanded + 1
        # for each action do
        for action in actions:
            # if child is not in explored or if cost is less
            if action.pos not in explored or action.cost < explored[action.pos].cost:
                # check if child is goal state
                if look(maze, action.pos) == "G":
                    path = action.get_path_from_root()
                    update_maze(maze, path)
                    return path, [n.cost for n in path][-1], nodes_expanded, max_t_depth, max_f_size
                # add child to frontier
                frontier.append(action)
                explored[action.pos] = action
                
        #update explored for maze
        if(look(maze, curr_node.pos) != "S" and look(maze, curr_node.pos) != "G"):
            maze[curr_node.pos] = "."
        
        if max_t_depth < len(curr_node.get_path_from_root()): 
            max_t_depth = len(curr_node.get_path_from_root())
            
        if max_f_size < len(frontier): 
            max_f_size = len(frontier)
                
    # if frontier got empty without returning path to goal state return none (failure)
    return None
```

```python
def dfs(maze):
    # store positions of initial and goal states
    start = find_pos(maze, what = "S")
    end = find_pos(maze, what = "G")
    # create node at position of initial state
    root_node = Node(pos = start, parent = None, action = None, cost = 0)
    # create stack to keep track of frontier
    frontier = []
    frontier.append(root_node)
    # create empty set for explored nodes
    explored = {}
    nodes_expanded = 0
    max_t_depth = 0
    max_f_size = 0
    
    #while frontier is not empty
    while frontier:
        # pop last node in the stack
        curr_node = frontier.pop(-1)
        # checks if node is in goal state position; if so return path
        if look(maze, curr_node.pos) == "G":
            path = curr_node.get_path_from_root()
            update_maze(maze, path)
            return path, [n.cost for n in path][-1], nodes_expanded, max_t_depth, max_f_size
        
        # add curr_node to explored set
        explored[curr_node.pos] = curr_node
        # get action/children nodes around curr_node
        actions = get_actions(maze, curr_node)
        nodes_expanded = nodes_expanded + 1
        # for each action do
        for action in actions:
            # if child is not in explored or if cost is less
            if action.pos not in explored or action.cost < explored[action.pos].cost:
                # check if child is goal state
                if look(maze, action.pos) == "G":
                    path = action.get_path_from_root()
                    update_maze(maze, path)
                    return path, [n.cost for n in path][-1], nodes_expanded, max_t_depth, max_f_size
                # add child to frontier
                frontier.append(action)
                explored[action.pos] = action
                
        #update explored for maze
        if(look(maze, curr_node.pos) != "S" and look(maze, curr_node.pos) != "G"):
            maze[curr_node.pos] = "."
        
        if max_t_depth < len(curr_node.get_path_from_root()): 
            max_t_depth = len(curr_node.get_path_from_root())
            
        if max_f_size < len(frontier): 
            max_f_size = len(frontier)
                
    # if frontier got empty without returning path to goal state return none (failure)
    return None
```

## Task 3: Implement greedy best-first search and A* search  [4 points]

You can use the map to estimate the distance from your current position to the goal using the Manhattan distance (see https://en.wikipedia.org/wiki/Taxicab_geometry) as a heuristic function. Both algorithms are based on Best-First search which requires only a small change from the BFS algorithm you have already implemented (see textbook/slides). 

```python
def heuristic_func(maze, start, end):
    x = np.absolute(start.pos[0] - end[0])
    y = np.absolute(start.pos[1] - end[1])
    return x + y

def gbfs(maze):
    #get start and end positions
    start = find_pos(maze, what = "S")
    end = find_pos(maze, what = "G")
    #initialize root node
    root_node = Node(pos = start, parent = None, action = None, cost = 0)
    #create frontier and add the root node
    frontier = []
    frontier.append(root_node)
    #other variables that are needed
    explored = {}
    nodes_expanded = 0
    max_t_depth = 0
    max_f_size = 0
    
    #while frontier is not empty
    while frontier:
        #keep track of heuristic values
        heuristic = []
        
        #iterate through nodes in frontier and get heur. values
        for node in frontier:
            heuristic.append(heuristic_func(maze, node, end))
        
        #find lowest heur value and its index
        low = min(heuristic)
        low_index = heuristic.index(low)
        
        #pop the node with the lowest heur. value
        curr_node = frontier.pop(low_index)
        
        #if curr_node is goal state then return
        if (look(maze, curr_node.pos) == "G"):
            path = curr_node.get_path_from_root()
            update_maze(maze, path)
            return path, [n.cost for n in path][-1], nodes_expanded, max_t_depth, max_f_size
        
        #if not, expand
        nodes_expanded += 1
        actions = get_actions(maze, curr_node)
        for action in actions:
            if action.pos not in explored or action.cost < explored[action.pos].cost:
                explored[action.pos] = action
                frontier.append(action)
                
                if maze[action.pos] != "S" and maze[action.pos] != "G":
                    maze[action.pos] = "."
                    
        #update explored for maze
        if(look(maze, curr_node.pos) != "S" and look(maze, curr_node.pos) != "G"):
            maze[curr_node.pos] = "."
        
        if max_t_depth < len(curr_node.get_path_from_root()): 
            max_t_depth = len(curr_node.get_path_from_root())
            
        if max_f_size < len(frontier): 
            max_f_size = len(frontier)
                
    # if frontier got empty without returning path to goal state return none (failure)
    return None
```

```python
def astar(maze):
    #get start and end positions
    start = find_pos(maze, what = "S")
    end = find_pos(maze, what = "G")
    #initialize root node
    root_node = Node(pos = start, parent = None, action = None, cost = 0)
    #create frontier and add the root node
    frontier = []
    frontier.append(root_node)
    #other variables that are needed
    explored = {}
    nodes_expanded = 0
    max_t_depth = 0
    max_f_size = 0
    
    #while frontier is not empty
    while frontier:
        #keep track of heuristic values
        heuristic = []
        
        #iterate through nodes in frontier and get heur. values
        for node in frontier:
            h = heuristic_func(maze, node, end)
            #add in node cost to heuristic value
            heuristic.append(h + node.cost)
        
        #find lowest heur value and its index
        low = min(heuristic)
        low_index = heuristic.index(low)
        
        #pop the node with the lowest heur. value
        curr_node = frontier.pop(low_index)
        
        #if curr_node is goal state then return
        if (look(maze, curr_node.pos) == "G"):
            path = curr_node.get_path_from_root()
            update_maze(maze, path)
            return path, [n.cost for n in path][-1], nodes_expanded, max_t_depth, max_f_size
        
        #if not, expand
        nodes_expanded += 1
        actions = get_actions(maze, curr_node)
        for action in actions:
            if action.pos not in explored or action.cost < explored[action.pos].cost:
                explored[action.pos] = action
                frontier.append(action)
                
                if maze[action.pos] != "S" and maze[action.pos] != "G":
                    maze[action.pos] = "."
                    
        #update explored for maze
        if(look(maze, curr_node.pos) != "S" and look(maze, curr_node.pos) != "G"):
            maze[curr_node.pos] = "."
        
        if max_t_depth < len(curr_node.get_path_from_root()): 
            max_t_depth = len(curr_node.get_path_from_root())
            
        if max_f_size < len(frontier): 
            max_f_size = len(frontier)
                
    # if frontier got empty without returning path to goal state return none (failure)
    return None
```

```python
# **THIS IS JUST MY DFS IMPLEMENTATION**
# IDS is just the DFS but iterated multiple times to increase the depth of the search
def ids(maze):
    # store positions of initial and goal states
    start = find_pos(maze, what = "S")
    end = find_pos(maze, what = "G")
    # create node at position of initial state
    root_node = Node(pos = start, parent = None, action = None, cost = 0)
    # create stack to keep track of frontier
    frontier = []
    frontier.append(root_node)
    # create empty set for explored nodes
    explored = {}
    nodes_expanded = 0
    max_t_depth = 0
    max_f_size = 0
    
    #while frontier is not empty
    while frontier:
        # pop last node in the stack
        curr_node = frontier.pop(-1)
        # checks if node is in goal state position; if so return path
        if look(maze, curr_node.pos) == "G":
            path = curr_node.get_path_from_root()
            update_maze(maze, path)
            return path, [n.cost for n in path][-1], nodes_expanded, max_t_depth, max_f_size
        
        # add curr_node to explored set
        explored[curr_node.pos] = curr_node
        # get action/children nodes around curr_node
        actions = get_actions(maze, curr_node)
        nodes_expanded = nodes_expanded + 1
        # for each action do
        for action in actions:
            # if child is not in explored or if cost is less
            if action.pos not in explored or action.cost < explored[action.pos].cost:
                # check if child is goal state
                if look(maze, action.pos) == "G":
                    path = action.get_path_from_root()
                    update_maze(maze, path)
                    return path, [n.cost for n in path][-1], nodes_expanded, max_t_depth, max_f_size
                # add child to frontier
                frontier.append(action)
                explored[action.pos] = action
                
        #update explored for maze
        if(look(maze, curr_node.pos) != "S" and look(maze, curr_node.pos) != "G"):
            maze[curr_node.pos] = "."
        
        if max_t_depth < len(curr_node.get_path_from_root()): 
            max_t_depth = len(curr_node.get_path_from_root())
            
        if max_f_size < len(frontier): 
            max_f_size = len(frontier)
                
    # if frontier got empty without returning path to goal state return none (failure)
    return None

extra_files = ["extra_maze1.txt", "extra_maze2.txt", "extra_maze3.txt"]

for file in extra_files:
    f = open(file, "r")
    maze_str = f.read()
    maze = parse_maze(maze_str)
    algo_title = ""
    for algo in range(len(algos) + 1):
        if algo == 0:
            algo_title = algos[algo]
            result = bfs(maze)
        elif algo == 1:
            algo_title = algos[algo]
            result = dfs(maze) 
        elif algo == 2:
            algo_title = algos[algo]
            result = gbfs(maze)
        elif algo == 3:
            algo_title = algos[algo]
            result = astar(maze)
        else:
            algo_title = "Iterative Deepening Search"
            result = ids(maze)
            
        print("Algorithm: ", algo_title)
        print("Path Cost: ", path_info[1])
        print("Nodes Expanded: ", path_info[2])
        print("Max Tree Size: ", path_info[3])
        print("Max Frontier Size: ", path_info[4])
        show_maze(maze)
```
